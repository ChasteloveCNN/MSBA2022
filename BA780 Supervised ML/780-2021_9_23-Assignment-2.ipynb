## Question 3

Is the following statement True or False?

> Average age of female athletes who attended the olympic games after 1990 has raised when compared to the era before then.

```python
# Your answer goes here
#After 1990:
female=athlete_events.loc[athlete_events['Sex']=='F']
female_a1990=female.loc[female['Year']>=1990]
print(female_a1990['Age'].mean())
```

```python
# Your answer goes here
#Before 1990:
female_b1990=female.loc[female['Year']<1990]
print(female_b1990['Age'].mean())
```
