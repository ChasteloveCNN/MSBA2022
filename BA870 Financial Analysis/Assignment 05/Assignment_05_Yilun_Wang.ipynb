### Run fixed effect regressions (using industry fixed effects) to explain companies' stock returns in early 2022 (RetYTD):

```python
# add constant column to the original dataframe
df['constant'] = 1
```

```python
df.columns
```

```python
# define x as a subset of original dataframe
x = df.drop(columns=['gvkey', 'datadate','fyear','indfmt','consol','popsrc','datafmt','Ticker','curcd','costat','Name ','RetYTD','constant'])
# define y as a series
y = df['RetYTD']

# pass x as a dataframe, while pass y as a series
model = sm.OLS(y, x).fit()
model.summary()
```
